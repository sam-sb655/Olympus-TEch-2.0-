```python
#Create a bar plot showing the count of wines for each quality score.
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\soumy\\Downloads\\PS_dataset.csv')
qscore=df['quality'].unique()
cow=df['quality'].value_counts()
cow1=cow.values.tolist()
plt.bar(qscore, cow1, color='darkred')
plt.xlabel('Quality score')
plt.ylabel('Count of wines')
plt.title('Bar plot')
plt.show()
```

```python
#Filter wines with a pH greater than 3.5 and alcohol content greater than 10. What’s the average quality of these wines?
import pandas as pd
df = pd.read_csv('C:\\Users\\soumy\\Downloads\\PS_dataset.csv')
filtered_wines = df[(df['pH'] > 3.5) & (df['alcohol'] > 10)]
average_quality = filtered_wines['quality'].mean()
print("Average quality of filtered wines:", average_quality)
```

```python
#Which quality score has the highest average alcohol content?
import pandas as pd
df = pd.read_csv('C:\\Users\\soumy\\Downloads\\PS_dataset.csv')
quality_grouped = df.groupby('quality')['alcohol'].mean()
highest_avg_alcohol_quality = quality_grouped.idxmax()
print("Quality score with highest average alcohol content:", highest_avg_alcohol_quality)
```

```python
#Group the dataset by quality and find the mean and median values for all other columns.
import pandas as pd
df = pd.read_csv('C:\\Users\\soumy\\Downloads\\PS_dataset.csv')
grouped_stats = df.groupby('quality').agg(['mean', 'median'])
print(grouped_stats)
```

```python
#How many null values are there in the citric acid column? Replace them with the median value of that column.
import pandas as pd
df = pd.read_csv('C:\\Users\\soumy\\Downloads\\PS_dataset.csv')
null_citric_acid = df['citric acid'].isnull().sum()
print("Null values in citric acid:", null_citric_acid)
df['citric acid'].fillna(df['citric acid'].median(), inplace=True)
```

```python
#Remove all rows where the residual sugar column has null values.
import pandas as pd
df = pd.read_csv('C:\\Users\\soumy\\Downloads\\PS_dataset.csv')
df.dropna(subset=['residual sugar'], inplace=True)
```

```python
#Normalize the fixed acidity column to a range between 0 and 1.
import pandas as pd
df = pd.read_csv('C:\\Users\\soumy\\Downloads\\PS_dataset.csv')
df['fixed acidity'] = (df['fixed acidity'] - df['fixed acidity'].min()) / (df['fixed acidity'].max() - df['fixed acidity'].min())
```
